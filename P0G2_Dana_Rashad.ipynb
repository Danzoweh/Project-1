#### Menghapus Whitespace

```python
import pandas as pd

df = df.rename(lambda x: x.strip() if isinstance('columns' , int) else x)

df
```

```python
import pandas as pd 

df = df.rename(columns=lambda x: x.strip())

df.info()
```

```python
df['Product'] = df['Product'].apply(lambda x: x.strip())

print(df)
```

```python
import pandas as pd 

df["Product"].unique()
```

```python
df['Discount Band'] = df['Discount Band'].apply(lambda x: x.strip())

df["Discount Band"].unique()
```

```python
df['Sales'] = df['Sales'].str.replace('\W', '', regex=True)
df['Profit'] = df['Profit'].str.replace('\W', '', regex=True)
df['COGS'] = df['COGS'].str.replace('\W', '', regex=True)
df['Units Sold'] = df['Units Sold'].str.replace('\W', '', regex=True)
df['Gross Sales'] = df['Gross Sales'].str.replace('\W', '', regex=True)
df['Sale Price'] = df['Sale Price'].str.replace('\W', '', regex=True)
df['Manufacturing Price'] = df['Manufacturing Price'].str.replace('\W', '', regex=True)
df['Discounts'] = df['Discounts'].str.replace('\W', '', regex=True)

df
```

### 3.Cek dan handling missing value. (Hint: terdapat non-standard missing value pada kolom yang semestinya bertipe data integer, handling dapat dilakukan dengan imputasi dengan angka nol '0')



```python
main_table = df

for column in main_table.columns:
    print(main_table[column].value_counts().head(), '\n')
```

```python
# Memisahkan data menjadi dua tabel untuk mencapai 2NF

# Tabel Pelanggan (Customer)
df_productsold = df[['Product', 'Units Sold']].drop_duplicates().reset_index(drop=True)

df_countrysales = df[['Country', 'Sales']].drop_duplicates().reset_index(drop=True)



print("\nTabel Pelanggan setelah Normalisasi ke 2NF:")
print(df_productsold)

print(df_countrysales)
```

```python
# Tal Pesanan (Order) tanpa informasi produk lainnya 

df_Countrysell = df[['Country','Product','Units Sold']].drop_duplicates().reset_index(drop=True)

print("\nTabel Pesanan setelah Normalisasi ke 3NF:")
print(df_Countrysell)
```
