# Data Cleaning

## Column Manipulation

### Penghapusan Column

```python
#hapus kolom month number , month name , dan year

df = df.drop(['Month Number', ' Month Name ', 'Year'], axis=1)
```

### Penggantian Nama Column

```python
df = df.rename(columns={" Product ":"Product"," Discount Band ":"Discount_Band",
                        " Units Sold ":"Units_Sold"," Manufacturing Price ":"Manufacturing_Price"," Sale Price ":"Sale_Price",
                        " Gross Sales ":"Gross_Sales"," Discounts ":"Discounts"," Sales ":"Sales"," COGS ":
                        "COGS"," Profit ":"Profit"," Month Number ":"Month_Number"," Month Name ":"Month_Name",})
```

```python

df = df.rename(columns={"Date": "Tanggal"})
```

```python
df
```

## Penambahan Column

```python
# Tambahkan kolom 'Order_ID' pada DataFrame
df.insert(0, 'Order_ID', range(1, 1 + len(df)))

df
```

## Data Changing

```python
#Menghilangkan White space pada column
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df
```

```python
#Hapus Dollar

for column in df.columns:
  if df[column].dtype == 'object':
    df[column] = df[column].str.replace('$', '', regex=False)

df

```

```python
#remove dot and comas from the column

for column in df.columns:
  if df[column].dtype == 'object':
    df[column] = df[column].str.replace('.', '', regex=False)
    df[column] = df[column].str.replace(',', '', regex=False)

df
```

```python
#ganti tipe data column Units_Sold , Manufacturing_Price , Sale_Price , Gross_Sale , Discounts , Sales , COGS dan Profit menjadi tipe data float

import pandas as pd
columns_to_convert = ['Units_Sold', 'Manufacturing_Price', 'Sale_Price', 'Gross_Sales', 'Discounts', 'Sales', 'COGS', 'Profit']

for column in columns_to_convert:
  df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
#ganti tipe data column Profit dan Discounts ke Integer
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce').fillna(0).astype(int)
df['Discounts'] = pd.to_numeric(df['Discounts'], errors='coerce').fillna(0).astype(int)
```

```python
#Ganti tipe data column Date menjadi Datetime64
df['Tanggal'] = pd.to_datetime(df['Tanggal'])

df.dtypes
```

```python
df.columns
```

## Check and Handling Missing Value

###Check Missing Value

```python
# Check for missing values in the DataFrame
df.isnull().sum()
```

###Handling Missing Value

```python
# Replace missing values in the 'Discounts' column with 0
df['Discounts'].fillna(0, inplace=True)

# Check if missing values have been handled
print(df.isnull().sum())
```

```python
# Interpolate missing values in the 'Profit' column
df['Profit'] = df['Profit'].interpolate()

# Check for missing values again to ensure interpolation worked
print(df.isnull().sum())

df
```

#Data Normalization

##Data Normalisasi sampai 3NF

```python
# Memisahkan tabel Product
product_df = df[['Product', 'Manufacturing_Price']].drop_duplicates()
product_df = product_df.rename(columns={'Manufacturing_Price': 'Product_Manufacturing_Price'})
product_df.insert(0, 'Product_ID', range(1, 1 + len(product_df)))

# Memisahkan tabel Segment
segment_df = df[['Segment']].drop_duplicates()
segment_df.insert(0, 'Segment_ID', range(1, 1 + len(segment_df)))

# Memisahkan tabel Country
country_df = df[['Country']].drop_duplicates()
country_df.insert(0, 'Country_ID', range(1, 1 + len(country_df)))

# Memisahkan tabel Discount Band
discount_band_df = df[['Discount_Band']].drop_duplicates()
discount_band_df.insert(0, 'Discount_Band_ID', range(1, 1 + len(discount_band_df)))

# Menggabungkan kembali tabel dengan foreign key
df = pd.merge(df, product_df, on='Product', how='left')
df = pd.merge(df, segment_df, on='Segment', how='left')
df = pd.merge(df, country_df, on='Country', how='left')
df = pd.merge(df, discount_band_df, on='Discount_Band', how='left')

# Membuat tabel Sales
sales_df = df[['Order_ID', 'Segment_ID', 'Country_ID', 'Product_ID', 'Discount_Band_ID', 'Units_Sold',
              'Sale_Price', 'Gross_Sales', 'Discounts', 'Sales', 'COGS', 'Profit', 'Tanggal']]

# Menghapus kolom yang tidak diperlukan di tabel sales
df = df.drop(['Product', 'Segment', 'Country', 'Discount_Band', 'Manufacturing_Price'], axis=1)

# Menampilkan tabel
print("Tabel Sales:")
print(sales_df.head())
print("\nTabel Product:")
print(product_df.head())
print("\nTabel Segment:")
print(segment_df.head())
print("\nTabel Country:")
print(country_df.head())
print("\nTabel Discount Band:")
print(discount_band_df.head())

# Simpan tabel ke file CSV
sales_df.to_csv('sales.csv', index=False)
product_df.to_csv('product.csv', index=False)
segment_df.to_csv('segment.csv', index=False)
country_df.to_csv('country.csv', index=False)
discount_band_df.to_csv('discount_band.csv', index=False)
```

##Primary Key dan Foreign Key

```python
"""
 Primary Keys:

 - segment_table: Segment_ID
 - country_table: Country_ID
 - product_table: Product_ID
 - discount_band_table: Discount_Band_ID
 - sales_table: Order_ID


 Foreign Keys:

 - sales_table:
   - Segment_ID referencing segment_table(Segment_ID)
   - Country_ID referencing country_table(Country_ID)
   - Product_ID referencing product_table(Product_ID)
   - Discount_Band_ID referencing discount_band_table(Discount_Band_ID)
"""
```

```python
# Primary key for Sales Table is a composite key consisting of:
# - Date
# - Segment_ID
# - Country_ID
# - Product_ID
# - Discount_Band_ID

# This means that a unique combination of these columns will identify a specific sales record.
```
